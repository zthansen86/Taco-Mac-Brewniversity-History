```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy.stats import sem
```

```python
# Read CSV files
beerlist_db = "Resources/beer_history.csv"

# Read file and put in dataframe
beerclub_db = pd.read_csv(beerlist_db)

beerclub_db.head()
```

```python
#Rename Columns
beerclub_db = beerclub_db.rename(columns={'Beer Name' : 'Beer',
                                          'Store Name' : 'Location',
                                          'Abv' : 'ABV',
                                          'Format' : 'Pour'})
beerclub_db.head()
```

```python
#Drop Unneccessary Columns
beerclub_db.drop(columns=['Beer PLU'], axis=1, inplace=True)
beerclub_db.head()
```

```python
# Collect a list of all the unique values in "Style"
beerclub_db["Style"].unique()
```

```python
# Create a group based on the values in the 'Style' column
style_group = beerclub_db.groupby('Style')

# Count how many times each style appears in our group
count_styles =style_group['Style'].count()

count_styles
```

```python
# Create a bar chart based off of the group series from before
count_chart = count_styles.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Style")
count_chart.set_ylabel("Number of Beers")


plt.show()
plt.tight_layout()
```

```python
# Filter the DataFrame down only to those columns to chart
beer_and_abv = beerclub_db[["Beer","ABV"]]

# Set the index to be "State" so they will be used as labels
beer_and_abv = beer_and_abv.set_index("Beer")

beer_and_abv.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
beer_and_abv.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("ABV Per Beer")

plt.show()
plt.tight_layout()
```
